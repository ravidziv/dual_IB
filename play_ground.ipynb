```python
df = pd.read_csv(path)
df = df.iloc[:step, :]
```

```python
inf_di = {'dual_x':'test_i_x_t_{}_{}_dual', 'dual_y':'test_i_y_t_{}_{}_dual', 'mine_x':'test_i_x_t_{}_mine', 'mine_y':'test_i_y_t_{}_mine',
        'bins_x':'testi_x_t_{}_{}_bins', 'bins_y':'test_i_y_t_{}_{}_bins','nlc_x':'test_i_x_t_{}_{}_c_nonlinear', 'nce_c_y':'test_i_y_t_{}_{}_c_nce'}
```

```python
fig, axes = plt.subplots(2,1, figsize = (10,8), sharex = True)
for  num_of_clusters_index in range(16,17):
    for layer_index in range(2,3 ):
        axes[0].plot(df['step'], df[inf_di['nlc_x'].format(layer_index,num_of_clusters_index )])
        axes[1].plot(df['step'], df[inf_di['nce_c_y'].format(layer_index,num_of_clusters_index )])

axes[0].set_xscale('log')
#plt.yscale('log')
```
